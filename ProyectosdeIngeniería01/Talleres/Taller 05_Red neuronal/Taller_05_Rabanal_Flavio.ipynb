## Importando librerías

```python
from keras.datasets import mnist
from keras import layers, models
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

## Nuestro set de datos

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

## Creando un modelo de datos

```python
model = models.Sequential()
model.add(layers.Dense(512,activation='relu', input_shape=(28*28,)))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Limpieza de datos

```python
x_train = train_data.reshape((60000,28*28))
x_train = x_train.astype('float32')/255

x_test = test_data.reshape((10000,28*28))
x_test = x_test.astype('float32')/255
```

```python
y_train = to_categorical(train_labels)
y_test = to_categorical(test_labels)
```

```python
train_labels[85]
```

```python
y_train[85]
```

## Entrenando el modelo

```python
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

## Evaluando sobre data de test

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

print("Calculando predicciones para la matriz...")
predicciones_prob = model.predict(x_test)
predicciones_clases = np.argmax(predicciones_prob, axis=1)

y_test_simple = np.argmax(y_test, axis=1)

matriz = confusion_matrix(y_test_simple, predicciones_clases)

plt.figure(figsize=(10, 8))
sns.heatmap(matriz, annot=True, fmt='d', cmap='Greens',
            xticklabels=range(10), yticklabels=range(10))

plt.title('Matriz de Confusión: ¿Qué números confunde la red?')
plt.xlabel('Predicción de la Red')
plt.ylabel('Número Real (Verdad)')
plt.show()

print("\n--- Reporte de Clasificación ---")
print(classification_report(y_test_simple, predicciones_clases))
```
