## Importando librerías

```python
from keras.datasets import mnist
from keras import layers, models
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

## Nuestro set de datos

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

## Creando un modelo de datos

```python
model = models.Sequential()
model.add(layers.Dense(512,activation='relu', input_shape=(28*28,)))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Limpieza de datos

```python
x_train = train_data.reshape((60000,28*28))
x_train = x_train.astype('float32')/255

x_test = test_data.reshape((10000,28*28))
x_test = x_test.astype('float32')/255
```

```python
y_train = to_categorical(train_labels)
y_test = to_categorical(test_labels)
```

```python
train_labels[26]
```

```python
y_train[26]
```

## Entrenando el modelo

```python
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

## Evaluando sobre data de test

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Hacemos que el modelo prediga todos los datos de prueba
y_pred = model.predict(x_test)

# Convertimos las probabilidades a números del 0 al 9
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Generamos y graficamos la matriz de confusión
cm = confusion_matrix(y_true_classes, y_pred_classes)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión')
plt.ylabel('Número Real')
plt.xlabel('Número Predicho por la Red')
plt.show()
```

```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. SOLICITAR ARCHIVOS
# Nota: En la ventana que se abra, puedes seleccionar varias imágenes a la vez
# manteniendo presionada la tecla 'Ctrl' (o 'Cmd' en Mac)
print("Selecciona tus imágenes (puedes elegir 2 o más):")
uploaded = files.upload()

# 2. PROCESAR CADA ARCHIVO SUBIDO
for fn in uploaded.keys():
    print(f"\n{'='*40}")
    print(f'PROCESANDO: "{fn}"')

    # --- PREPARACIÓN DE IMÁGENES ---
    # Imagen original para mostrar (convertida a RGB)
    img_color = cv2.imread(fn)
    img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

    # Imagen para la red (Escala de grises -> Inversión -> Umbral)
    img_gray = cv2.imread(fn, cv2.IMREAD_GRAYSCALE)
    img_inv = cv2.bitwise_not(img_gray)
    _, img_bin = cv2.threshold(img_inv, 120, 255, cv2.THRESH_BINARY)

    # Redimensionar a la escala de MNIST (28x28)
    img_final = cv2.resize(img_bin, (28, 28))

    # --- VISUALIZACIÓN LADO A LADO ---
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    # Lado izquierdo: Foto original
    axes[0].imshow(img_color)
    axes[0].set_title(f"Original: {fn}")
    axes[0].axis('off')

    # Lado derecho: Lo que recibe el modelo
    axes[1].imshow(img_final, cmap='gray')
    axes[1].set_title("Entrada para la Red (28x28)")
    axes[1].axis('off')

    plt.show()

    # --- PREDICCIÓN ---
    # Preparar el array: (1, 784) y normalizar
    img_array = img_final.reshape((1, 28 * 28))
    img_array = img_array.astype('float32') / 255

    prediccion = model.predict(img_array, verbose=0) # verbose=0 para limpiar la salida
    clase_predicha = np.argmax(prediccion)
    confianza = np.max(prediccion) * 100

    print(f" RESULTADO: La red neuronal dice que es un {clase_predicha}")
    print(f" CONFIANZA: {confianza:.2f}%")
    print(f"{'='*40}\n")
```
